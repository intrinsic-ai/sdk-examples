```python
from intrinsic.solutions import deployments
from intrinsic.solutions import behavior_tree as bt
```

First, connect to the deployed solution:

```python
solution = deployments.connect_to_selected_solution()
executive = solution.executive
skills = solution.skills
points_crud_skill = skills.com.example.points_crud
```

Store two points in the KV storage:

```python
create_points = points_crud_skill(create=True, return_value_key="create_points_log")
executive.run(create_points)

result = executive.get_value(create_points.result)
print(result)
```

## Redeploy your solution now

It will demonstrate that data in the key-value storage survives redeployment.

- In Flowstate, select File -> Redeploy.
- Wait for redeployment to complete before running code in cells below.

Reconnect to the solution after it has been redeployed:

```python
solution = deployments.connect_to_selected_solution()
executive = solution.executive
skills = solution.skills
points_crud_skill = skills.com.example.points_crud
```

Confirm that the contents of the KV storage survived the redeployment:

```python
read_points = points_crud_skill(read=True, return_value_key="read_log")
executive.run(read_points)
result = executive.get_value(read_points.result)
print(result)
```

Update a value stored in the KV storage:

```python
update_point = points_crud_skill(update=True, return_value_key="update_log")
executive.run(update_point)
result = executive.get_value(update_point.result)
print(result)
```

Finally, delete all points from the KV storage:

```python
delete_points = points_crud_skill(delete=True, return_value_key="delete_log")
executive.run(delete_points)
result = executive.get_value(delete_points.result)
print(result)
```
